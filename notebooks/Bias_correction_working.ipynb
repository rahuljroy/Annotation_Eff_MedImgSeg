```python
import numpy as np
from nipype.interfaces.ants import N4BiasFieldCorrection
import os
from multiprocessing import Pool
import pickle
```

```python
def N4ITK(filepath):
    print('Working on: ', filepath)
    n4 = N4BiasFieldCorrection()
    n4.inputs.dimension = 3
    n4.inputs.input_image = filepath
    
#     outputPath_train = '../data/Processed/test/'+filepath[-22:-7]+'_biascorrected.nii.gz'
    outputPath_test = '../data/Processed/test/'+filepath[-16:-7]+'_biascorrected.nii.gz'
    n4.inputs.output_image = outputPath_test
    
    n4.run()
```

```python
test_path = '../data/Task01_BrainTumour/imagesTr'
# training_path = '../data/Task01_BrainTumour/imagesTr'
```

```python
import glob
filepaths = glob.glob(test_path +"/*.nii.gz")
```
